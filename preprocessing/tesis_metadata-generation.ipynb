## Generate metadata files

```python
def part_code(part_text):
    return part_text.lower().replace(" ", "_")

def generate_metadata_data_from_row(row):
    result = {}
    result["complaint"] = row["DENUNCIA"]
    result["accident"] = row["STRO"]
    result["brand"] = row["TR_MARCA"].lower()
    result["model"] = row["TR_MODELO"].lower()
    result["type"] = row["TIPO_VEH"].lower()
    result["color"] = row["COLOR"].lower()
    result["damage_comment"] = row["TR_DANIOS"].lower()
    result["car_year"] = math.trunc(row["TR_ANIO"]*1000.0+1)

    result["damaged_parts"] = [{"part": part_code(x), "part_text": x, "task": "repair", "hours":0} for x in row["piezas_normalizadas"].split(',')]
    return result

def save_metadata_file(data):
    filename = "./{}/{}/metadata.json".format("metadata_files", data["complaint"])
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    with open(filename, "w") as f:
        json.dump(data, f)
```
